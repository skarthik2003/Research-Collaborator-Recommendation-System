````python
# Assuming df and data DataFrames have been loaded and columns renamed appropriately
df1 = df[df['Publication_Year'] <= 2018]

# Merge the datasets on 'Author_ID' to include research concepts for each author
merged_df = pd.merge(df1, data[['Author_ID', 'Research_Concepts']], on='Author_ID', how='left')

# Initialize a new graph (or you can update G_until_2022 if you prefer)
G_with_concepts = nx.Graph()
for i, row in merged_df.iterrows():
    if row['Author_ID'] not in G_with_concepts:
        G_with_concepts.add_node(row['Author_ID'],
                                 name=row['Author Name'],
                                 h_index=row['H-Index'],
                                 i_index=row['I-Index'],
                                 mean_citedness=row['Two-Year_Mean_Citedness'],
                                 institution=row['Institution'],
                                 country_code=row['Country_Code'],
                                 research_concepts=row['Research_Concepts'])
    if row['Co-Author_ID'] not in G_with_concepts:
        G_with_concepts.add_node(row['Co-Author_ID'], name=row['Co-Author Display Name'])
    if not G_with_concepts.has_edge(row['Author_ID'], row['Co-Author_ID']):
        G_with_concepts.add_edge(row['Author_ID'], row['Co-Author_ID'],
                                 publication_year=row['Publication_Year'],
                                 avg_max_cited_by_percentile=row['Citation'],
                                 weight=row['Citation'])
    else:
        current_weight = G_with_concepts[row['Author_ID']][row['Co-Author_ID']].get('weight', 0)
        G_with_concepts[row['Author_ID']][row['Co-Author_ID']]['weight'] = current_weight + row['Citation']

# Now G_with_concepts includes research concepts for authors
````

````python
# Since the actual data and graph creation are conceptual, this code assumes the graph G_with_concepts has been created as described.

plt.figure(figsize=(12, 10))
plt.title("Cumulative Co-Authorship Network until 2022")

# Random layout for better visualization of G_with_concepts
pos = nx.random_layout(G_with_concepts)

# Draw the graph with edge colors based on weights
edges = G_with_concepts.edges()
weights = [G_with_concepts[u][v]['weight'] for u, v in edges]

# Normalizing the weights for color mapping
norm = Normalize(vmin=min(weights), vmax=max(weights))
edge_colors = [plt.cm.Reds(norm(weight)) for weight in weights]

# Draw edges with colors and create a colorbar
nx.draw(G_with_concepts, pos, with_labels=False, node_size=50, node_color='skyblue', edge_color=edge_colors, width=2.0, edge_cmap=plt.cm.Reds)

# Create a ScalarMappable for colorbar
sm = ScalarMappable(cmap='Reds', norm=norm)  # Use a red colormap for the ScalarMappable
sm.set_array([])  # Set the range of values for the colorbar
cbar = plt.colorbar(sm, label='Edge Weight')  # Create the colorbar based on ScalarMappable
cbar.ax.yaxis.set_label_position('left')  # Adjust the colorbar label position

plt.show()
````

````python
from itertools import combinations
import networkx as nx

def jaccard_coefficient_prediction(G):
    new_edges = []
    for u, v in combinations(G.nodes(), 2):
        if not G.has_edge(u, v):
            # Calculate the Jaccard Coefficient
            union_size = len(set(G.neighbors(u)) | set(G.neighbors(v)))
            intersection_size = len(list(nx.common_neighbors(G, u, v)))
            if union_size > 0:  # Prevent division by zero
                jaccard_score = intersection_size / union_size
            else:
                jaccard_score = 0
            new_edges.append((u, v, jaccard_score))
    # Sort by score in descending order before returning
    new_edges_sorted = sorted(new_edges, key=lambda x: x[2], reverse=True)
    return new_edges_sorted

# Example usage
jaccard_preds = jaccard_coefficient_prediction(G_with_concepts)
print("Total predictions:", len(jaccard_preds))
for edge in jaccard_preds[:10]:
    print(edge)
````

````python
import networkx as nx
from itertools import combinations
import math

def adamic_adar_prediction(G):
    new_edges = []
    for u, v in combinations(G.nodes(), 2):
        if not G.has_edge(u, v):
            # Calculate the Adamic-Adar index
            common_neighbors = set(G.neighbors(u)) & set(G.neighbors(v))
            adamic_adar_index = sum(1 / math.log(G.degree(w)) for w in common_neighbors)

            # Calculate the probability using the Adamic-Adar index
            probability = 1 - math.exp(-adamic_adar_index)
            new_edges.append((u, v, probability))
    return new_edges

predicted_edges = adamic_adar_prediction(G_with_concepts)
predicted_edges = sorted(predicted_edges, key=lambda x: x[2], reverse=True)

# Print the top 10 predicted edges with their probabilities
print(len(predicted_edges))
for edge in predicted_edges[:10]:
    print(f"({edge[0]}, {edge[1]}) -> {edge[2]:.8f}")
````

````python
import numpy as np

def normalize_scores_and_compute_mean_log_and_minmax(score_lists, threshold=0, epsilon=1e-5):
    # Collect all unique node pairs
    node_pairs = set()
    for score_list in score_lists:
        node_pairs.update((u, v) for u, v, score in score_list if score > 0)

    # Prepare a dictionary to hold scores for each node pair
    scores_dict = {pair: [] for pair in node_pairs}

    # Gather scores for each pair across all lists
    for score_list in score_lists:
        for u, v, score in score_list:
            if score > 0:
                scores_dict[(u, v)].append(score)

    # Apply log scaling and compute the mean score for each node pair
    logged_means = {}
    for pair, scores in scores_dict.items():
        if scores:
            scores = np.array(scores)
            # Apply Log scaling: log(score + epsilon)
            logged_scores = np.log(scores + epsilon)
            mean_logged_score = np.mean(logged_scores)
            # Store the mean of logged scores if it's above 0
            if mean_logged_score > 0:
                logged_means[pair] = mean_logged_score

    # Normalize the logged means using Min-Max normalization
    normalized_scores = {}
    if logged_means:
        # Calculate min and max of logged means
        values = list(logged_means.values())
        min_value = min(values)
        max_value = max(values)

        # Apply Min-Max normalization
        for pair, score in logged_means.items():
            normalized_score = (score - min_value) / (max_value - min_value) if max_value > min_value else 0
            normalized_scores[pair] = normalized_score

    # Optionally print the normalized results
    print("Normalized Min-Max Scores for each Node Pair:")
    for pair, score in normalized_scores.items():
        print(f"{pair}: {score:.3f}")

    return normalized_scores

# Example usage
normalized_scores = normalize_scores_and_compute_mean_log_and_minmax([
    matches_preferential,
    matches_resource_allocation,
    matches_common_neighbors_prediction,
    matches_jaccard_preds,
    matches_adamic_adar_index
], threshold=-np.inf)  # Set to -inf to include all scores in the printout

# `normalized_scores` now holds the data normalized by both log scaling and Min-Max
````
